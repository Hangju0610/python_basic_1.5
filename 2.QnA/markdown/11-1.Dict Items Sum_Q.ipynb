````python
# 11. Dict Items Sum
d = {'a': 17,'b': 114,'c': 247, 'd': 362, 'e': 220, 'f': 728, 'g': -283, 'h': 922}

print(d.values())

# ë°©ë²•1
total = 0

for i in d.values():
    total += i

print(total)

# ë°©ë²•2
total2 = sum(d.values())
print(total2)

# ë°©ë²•3
total3 = sum(d[item] for item in d)
print(total3)

# ë°©ë²•4 (reduce)
# ....
````
